```python
#question:which year recorded the highest abortion rate
import pandas as pd
data=pd.read_csv("abortion-statistics-year-ended-december-2018-abortion-rates-by-age-of-woman-csv.csv")
maximum=data.groupby('Period').sum()
maximum
```

```python
#data visualization>>>Bar graph
import matplotlib.pyplot as plt
plt.bar(x=data['Period'],height=data['Abortion_rate'])
Period=range(2000,2018,2)
plt.xticks(Period)
plt.xlabel('Periods')
plt.ylabel('Abortion rate')
plt.show()
```

```python
#Question two:under which age bracket,was there recorded the highest abortion rate
result=data.groupby('Age_of_woman').sum()
result
```

```python
plt.bar(x=data['Age_of_woman'],height=data['Abortion_rate'])
plt.xlabel('Age bracket')
plt.ylabel('Abortion rate')
plt.title('Abortion rate as per Age bracket')
plt.show()
```
